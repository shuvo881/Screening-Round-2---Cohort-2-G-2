```python
import pandas as pd
import datetime
```

# Load the **raw_data.csv** file into a pandas DataFrame. 

---



---



---



---



```python
df = pd.read_csv('raw_data.csv')

df.info
```

# Check the data for **missing** values and **duplicated** records.

---



---



---






```python
print(df.isnull().sum()) #print sum of Null Fields in each column
```

```python
print("The Total number of Null Fields are: ", df.isnull().sum().sum())
```

```python
print("The Total Duplicated Records are :",df.duplicated().sum()) # 
```

# **Remove** any duplicate records from the data. 

---



---



---



---



```python
df = df.drop_duplicates()
```

```python
df.info
```

# Fill in **any missing values** in **the gender, marital_status, and city** columns with the **mode** of the respective columns. 

---



---



---



```python
# Calculate the mode of each column
mode_gender = df['gender'].mode()[0]
mode_marital_status = df['marital_status'].mode()[0]
mode_city = df['city'].mode()[0]

# Fill missing values with the respective mode
df['gender'] = df['gender'].fillna(mode_gender)
df['marital_status'] = df['marital_status'].fillna(mode_marital_status)
df['city'] = df['city'].fillna(mode_city)
```

```python
df.isnull().sum()
```

# Create a **new column** named **age** that contains the age of each customer based on their dob.

---



---



---



---



```python
# Convert the dob column to a datetime object
df['dob'] = pd.to_datetime(df['dob'])

# Calculate the age based on the current date
current_date = datetime.datetime.now()
df['age'] = df['dob'].apply(lambda x: (current_date - x).days // 365)
```

```python
df.info
```

# Create a new column named **income_group** that categorizes customers into three groups based on their income values: **"Low", "Medium", "High"** based on each **33% percentile**.

---



---



---



---



```python
df['income_group'] = pd.qcut(df['income'], q=[0, .33, .66, 1], labels=['Low', 'Medium', 'High'])
# low range = 0-0.33
# midum range = 0.33-0.66
# high range = 0.66-1
```

```python
df
```

# Create a new column named **score_group** that categorizes customers into three groups based on their **score** values: **"Poor", "Fair", and "Good"** based on each **33% percentile**. 

---



---



---






```python
df['score_group'] = pd.qcut(df['score'], q=[0, .33, .66, 1], labels=['Poor', 'Fair', 'Good'])
# Poor range = 0-0.33
# Fair range = 0.33-0.66
# Good range = 0.66-1
```

```python
df
```

# **Remove** any **rows** where the last_purchase_date is **before** the year **2019**

---



---



---



```python
# Convert last_purchase_date column to datetime format
df['last_purchase_date'] = pd.to_datetime(df['last_purchase_date'])

# Remove rows where last_purchase_date is before 2019
df = df[df['last_purchase_date'].dt.year >= 2019]
```
